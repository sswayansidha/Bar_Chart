# **Pie Chart**

```python
import pandas as pd
data = {'product_name': ['laptop', 'printer', 'tablet', 'desk', 'chair'],
        'price': [1200, 150, 300, 450, 200]
        }

df = pd.DataFrame(data)
print (df)
```

```python
from matplotlib import pyplot as plt
```

```python
Product = list(df.product_name)
Price = list(df.price)
fig = plt.figure(figsize=(10,7))
plt.pie(Price,labels=Price)
#plt.pie(Price,labels=Price)
plt.show()
```

**With % & Product Level**

```python
import matplotlib.pyplot as plt
%matplotlib inline

Product = list(df.product_name)
Price = list(df.price)

colors = ['lightskyblue', 'red', 'blue', 'green', 'gold']
plt.pie(Price, labels=Product , colors=colors, startangle=90, autopct='%.1f%%')
plt.show()
```

**Explode Pie**

```python
## for exploding the pie chart 
Product = list(df.product_name)
Price = list(df.price)

colors = ['lightskyblue', 'red', 'blue', 'green', 'gold']
explode =(0,0.1,0,0.1,0)
plt.pie(Price, labels=Product ,explode = explode, colors=colors, startangle=90, autopct='%.1f%%')
plt.show()
```

**Adding Shadows and title to explode pie**

```python
## Adding a title to the pie chart and a shadow to the chart 
Product = list(df.product_name)
Price = list(df.price)

colors = ['lightskyblue', 'red', 'blue', 'green', 'gold']
explode =(0,0.1,0,0.1,0)
plt.pie(Price, explode = explode,labels= Product, colors=colors, startangle=90, autopct='%.1f%%' ,shadow = True)
plt.title("Product Price Distribution", fontsize = 18) 
plt.show()
```

**Adding Legends to Pie chart**

```python
## Adding a legend to the chart and increasing its size 
Product = list(df.product_name)
Price = list(df.price)

colors = ['lightskyblue', 'red', 'blue', 'green', 'gold']
explode =(0,0.1,0,0.1,0)
fig, ax1 = plt.subplots(figsize = (12,7))
plt.pie(Price, explode = explode,labels= Product, colors=colors, startangle=90, autopct='%.1f%%' ,shadow = True)
plt.title("Product Price Distribution", fontsize = 18) 
ax1.legend(Product, loc = "upper right",fontsize = 10) 
plt.tight_layout() 
plt.show()
```

**Labeling Exact values to pie**

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 5), subplot_kw=dict(aspect="equal"))

Product = list(df.product_name)
Price = list(df.price)

def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)


wedges, texts, autotexts = ax.pie(Price, autopct=lambda pct: func(pct, Price),
                                  textprops=dict(color="w"))

ax.legend(wedges, Product,
          title="Product",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("A pie")

plt.show()
```

**Border to Pie Chart**

```python
Product = list(df.product_name)
Price = list(df.price)

#colors = ['lightskyblue', 'red', 'blue', 'green', 'gold']
fig, ax1 = plt.subplots(figsize = (5,12))
plt.pie(Price,labels= Product, startangle=90, autopct='%.1f%%' ,wedgeprops = {"edgecolor" : "black",'linewidth': 2,'antialiased': True})

plt.title("Product Price Distribution", fontsize = 18) 
ax1.legend(Product, loc = "upper right",fontsize = 10) 
plt.tight_layout() 
plt.show()
```
